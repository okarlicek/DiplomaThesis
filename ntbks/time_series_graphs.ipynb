```julia
using Plots
using StatsPlots
using Random
include("../src/utils/fcn_data.jl");
```

# AR2

```julia
Random.seed!(1)

theta = [0.2, -0.9]
setup = Dict("mod" => Dict("model" => "AR2",
                           "fixed_params" => [sqrt(0.1)]))

data = generate_series(setup, 3000, 500, theta);

ar2_plot = plot(data, linecolor=:black, legend=false, xlabel="time", ylabel="series", size=(800, 400), margin=3Plots.mm)
```

```julia
savefig(ar2_plot, "../plots/ar2_plot_realizations.png")
```

# FW 2012

```julia
Random.seed!(1)

theta = [0.12, 1.50, - 0.327, 1.79, 18.43]
setup = Dict("mod" => Dict("model" => "fw2012_5p",
                           "fixed_params" => [0.01, 0.758, 2.087, 0.0, 1.]))

data = generate_series(setup, 3000, 500, theta);

fw_plot_realization = plot(data, linecolor=:black, legend=false, xlabel="time", ylabel="series", size=(800, 400), margin=3Plots.mm)
```

```julia
savefig(fw_plot_realization, "../plots/fw_plot_realization.png")
```

```julia
fw_plot_differences = plot(diff(data), linecolor=:black, legend=false, xlabel="time", ylabel="series", size=(800, 400), margin=3Plots.mm)
```

```julia
savefig(fw_plot_differences, "../plots/fw_plot_differences.png")
```
